```python
#cau2
#a
import numpy as np
import math
n=int(input("nhap so tu nhien n duong:"))
b=[]
for i in np.arange(0,n):
  b.append(12+i*2)
print("cau a: b=",b)
#b
c=[]
for i in np.arange(0,n):
  c.append(31+i*2)
print("cau b: c=",c)
#c
x= []
for i in range (0,2*n+1):
  x.append(i-n)
print("cau c x= ",x)
#d
d = []
for i in range (0,2*n+1):
  d.append(n-i)
print("cau d y= ",d)
#e
e=[]
for i in np.arange(-n/2-4,n/2):
  if(i%2==0):
       e.append(-int(i))
print("cau e: z=",e)
#f
f=[]
for i in np.arange(0,n):
  if(i%2==0):
      if(i==0):
          f.append(1)
      else:
          f.append(1/int(i))
print("cau f: w=",f)
#g
fibo=[1,1]
g=[]
for i in np.arange(2,n):
  fibo.append(fibo[i-2]+fibo[i-1])
print(fibo)
for i in np.arange(0,n):
    g.append(1/fibo[i])
print("cau g: d=",g)
#h
h = []
count = 0
num = 2
while(count<n):
    for i in np.arange(2, num+1):
        if num == 2:
            h.append(1/num)
            num = num + 1
            count = count + 1
            continue
        if num % i == 0 and num != i :
            num = num + 1
            continue
    h.append(1/num)
    num = num + 1
    count = count + 1
print("cau h: e=",h)
#i
i1 = []
i1.append(1)
sum = 1
for i in np.arange(1, n+1):
    sum = sum + (i+1)
    i1.append(sum)
print("a = ", i1)
#jj
j1=[]
for i in np.arange(1,n):
  j1.append(1/(i**2+1))
print("n=",j1)
#k
p = []
num1 = 0
num2 = 1
for i in np.arange(n):
    p.append((num1+i)/(num2+i))
print("p = ", p)
#caul
#ord()
#chr()
char=[]
for i in np.arange(ord('a'),ord('z')+1):
  char.append(chr(i))
print('char=',char)
#caum
CHAR=[]
for i in np.arange(ord('A'),ord('Z')+1):
  CHAR.append(chr(i))
print('char=',CHAR)
```
