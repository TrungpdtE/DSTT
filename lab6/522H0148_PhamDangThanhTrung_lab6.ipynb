```python
#cau4
import numpy as np
#a
u=np.array([1,1])
v=np.array([0,1])
normu=np.linalg.norm(u)
normv=np.linalg.norm(v)
print("a. the angle between u and v:", round(np.degrees(np.arccos(np.dot(u,v)/(normu*normv))),5) )
#b
u=np.array([1,0])
v=np.array([0,1])
normu=np.linalg.norm(u)
normv=np.linalg.norm(v)
print("b. the angle between u and v:", round(np.degrees(np.arccos(np.dot(u,v)/(normu*normv))),5) )
#c
u=np.array([-2,3])
v=np.array([1/2,-1/2])
normu=np.linalg.norm(u)
normv=np.linalg.norm(v)
print("c. the angle between u and v:", round(np.degrees(np.arccos(np.dot(u,v)/(normu*normv))),5) )
```

```python
#cau5
#a
u=np.array([2,3])
normu=np.linalg.norm(u)
print("A. norm: ",normu)
Unitvecto=u/normu
print("A. Unit vecto u: ",Unitvecto)
print()
#b
u=np.array([1,2,3])
normu=np.linalg.norm(u)
print("B. norm: ",normu)
Unitvecto=u/normu
print("B. Unit vecto u: ",Unitvecto)
print()
#c
u=np.array([1/2,-1/2,1/4])
normu=np.linalg.norm(u)
print("C. norm: ",normu)
Unitvecto=u/normu
print("C. Unit vecto u: ",Unitvecto)
print()
#d
u=np.array([2**(1/2),2,-(2**(1/2)),2**(1/2)])
normu=np.linalg.norm(u)
print("D. norm: ",normu)
Unitvecto=u/normu
print("D. Unit vecto u: ",Unitvecto)
print()
```

```python
#cau6
import numpy as np
v1=np.array([1,2,3])
s2=np.array([7,4,3])
s3=np.array([2,1,9])
eucli= np.sqrt(np.dot((v1-s2).T,v1-s2 ))
print("Euclidean distances v1-s2 :",eucli)

eucli= np.sqrt(np.dot((v1-s3).T,v1-s3 ))
print("Euclidean distances v1-s3 :",eucli)

eucli= np.sqrt(np.dot((s2-s3).T,s2-s3 ))
print("Euclidean distances s2-s3 :",eucli)
```
