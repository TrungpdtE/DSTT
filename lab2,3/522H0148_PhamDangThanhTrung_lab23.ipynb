```python
#cau10
import numpy as np
A=np.array([[-1,4,8],[-9,1,2]])
B=np.array([[5,8],[0,-6],[5,6]])
C=np.array([[-4,1],[6,5]])
D=np.array([[-6,3,1],[8,9,-2],[6,-1,5]])
print('A=',A)
print('B=',B)
print('C=',C)
print('D=',D)
print('a) AB^T khong ton tai')
print('b) BC^T=',np.matmul(B,C.T))
print('c)(C-C^T)=',C-C.T)
print('d)(D-D^T)=',D-D.T)
print('e) (D^T)^T=',(D.T).T)
print('f) 2C^T=',2*(C.T))
print('g) A^T+B=',A.T+B)
print('h)(A^T+B)^T=',(A.T+B).T)
print('((2A^T-5B)^T=',(2*A.T-5*B).T)
print('((-D)^T)=',(-1*D).T)
print('(-D^T)=',-1*(D.T))
print('((C^2)^T=',(C*2).T)
print('(C^T)^2=',(C.T)*(C.T))
```

```python
#cau15
import numpy as np
a=np.array([[2,4,5/2],[-3/4,2,1/4],[1/4,1/2,2]])
b=np.array([[1,-1/2,3/4],[3/2,1/2,-2],[1/4,1,1/2]])
#a^-1
a_inv=np.linalg.inv(a)
b_inv=np.linalg.inv(b)

print("15a.1\n",np.matmul(a_inv,b_inv))
print("15a.2\n",np.linalg.inv(np.matmul(a,b)))
print("15a.3\n",np.linalg.inv(np.matmul(b,a)))

print("15b.1\n",(a_inv).T)
print("15b.2\n",np.linalg.inv(a.T))
```

```python
#cau13

import numpy as np

def random_matrix(n):
    return np.random.rand(n, n)

#chaythumatran5x5
A = random_matrix(5)
B = random_matrix(5)
det_A = np.linalg.det(A)
det_B = np.linalg.det(B)
det_A_plus_B = np.linalg.det(A + B)

print("chaythumatran5x5")
print("det(A + B) - det(A) - det(B) = ", det_A_plus_B - det_A - det_B)
print("Is det(A + B) = det(A) + det(B)?", np.allclose(det_A_plus_B, det_A + det_B))

#chaythumatran3x3
A = random_matrix(3)
B = random_matrix(3)
det_A = np.linalg.det(A)
det_B = np.linalg.det(B)
det_A_plus_B = np.linalg.det(A + B)

print("chaythumatran3x3")
print("det(A + B) - det(A) - det(B) = ", det_A_plus_B - det_A - det_B)
print("Is det(A + B) = det(A) + det(B)?", np.allclose(det_A_plus_B, det_A + det_B))

#chaythumatran4xx4
A = random_matrix(4)
B = random_matrix(4)
det_A = np.linalg.det(A)
det_B = np.linalg.det(B)
det_A_plus_B = np.linalg.det(A + B)

print("chaythumatran4xx4")
print("det(A + B) - det(A) - det(B) = ", det_A_plus_B - det_A - det_B)
print("Is det(A + B) = det(A) + det(B)?", np.allclose(det_A_plus_B, det_A + det_B))

#chaythumatran2xx2
A = random_matrix(2)
B = random_matrix(2)
det_A = np.linalg.det(A)
det_B = np.linalg.det(B)
det_A_plus_B = np.linalg.det(A + B)

print("chaythumatran2xx2")
print("det(A + B) - det(A) - det(B) = ", det_A_plus_B - det_A - det_B)
print("Is det(A + B) = det(A) + det(B)?", np.allclose(det_A_plus_B, det_A + det_B))
```

```python
#cau14

import numpy as np

def random_matrix(n):
    return np.random.rand(n, n)

#chaythumatran2xx2
A = random_matrix(2)
B = random_matrix(2)
det_A = np.linalg.det(A)
det_B = np.linalg.det(B)
det_AB = np.linalg.det(A.dot(B))

print("chaythumatran2xx2")
print("det(AB) - det(A) * det(B) = ", det_AB - det_A * det_B)
print("Is det(AB) = det(A) * det(B)?", np.allclose(det_AB, det_A * det_B))

#chaythumatran3x3
A = random_matrix(3)
B = random_matrix(3)
det_A = np.linalg.det(A)
det_B = np.linalg.det(B)
det_AB = np.linalg.det(A.dot(B))

print("chaythumatran3x3")
print("det(AB) - det(A) * det(B) = ", det_AB - det_A * det_B)
print("Is det(AB) = det(A) * det(B)?", np.allclose(det_AB, det_A * det_B))

# chaythumatran44
A = random_matrix(4)
B = random_matrix(4)
det_A = np.linalg.det(A)
det_B = np.linalg.det(B)
det_AB = np.linalg.det(A.dot(B))

print("chaythumatran4xx4")
print("det(AB) - det(A) * det(B) = ", det_AB - det_A * det_B)
print("Is det(AB) = det(A) * det(B)?", np.allclose(det_AB, det_A * det_B))

#chaythumatran5xx5
A = random_matrix(5)
B = random_matrix(5)
det_A = np.linalg.det(A)
det_B = np.linalg.det(B)
det_AB = np.linalg.det(A.dot(B))

print("chaythumatran5xx5")
print("det(AB) - det(A) * det(B) = ", det_AB - det_A * det_B)
print("Is det(AB) = det(A) * det(B)?", np.allclose(det_AB, det_A * det_B))
```
